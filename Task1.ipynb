```python
# obtain the attrition rate
data['Attrition'].value_counts(normalize=True)
```

```python
attrition = data['Attrition'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,6))
ax = sns.barplot(x=attrition.index, y=attrition)
for p in ax.patches:
    ax.annotate(f'{p.get_height() * 100:.2f}%',

    (p.get_x() + p.get_width() / 2., p.get_height()),
    ha='center', va='bottom')
plt.title('Distribution of Attrition Rate')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.tight_layout()
plt.show()
```

```python
avg_tenure = data['YearsAtCompany'].mean()
print(f'Average years of employee to leave the company is {avg_tenure} years')
```

The barplot shows that the company's attrition rate is 16.12% which means that 16.12% people have left the company within the period of averagely 7 years.

```python
sns.histplot(data, x='Age', kde=True)
```

The graph show that the maximum people working in the company are in the range of 27-40 years

```python
# separated the numerical and the categorical data
num_data = data.select_dtypes(include=['int64', 'float64'])
cat_data = data.select_dtypes(include=['object', 'category'])
```

```python
num_data.columns
```

```python
# obtained the correlation matrix
corr_matrix = num_data.corr().round(2)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# dropped the columns having high correlation
num_data.drop("JobLevel",axis=1,inplace=True)
num_data.drop("PerformanceRating",axis=1,inplace=True)
num_data.drop("TotalWorkingYears",axis=1,inplace=True)
num_data.drop("YearsAtCompany",axis=1,inplace=True)
num_data.drop("YearsInCurrentRole",axis=1,inplace=True)
num_data.drop("YearsSinceLastPromotion",axis=1,inplace=True)
num_data.drop("YearsWithCurrManager",axis=1,inplace=True)
num_data.drop("EmployeeNumber",axis=1,inplace=True)
num_data.drop("StandardHours",axis=1,inplace=True)
```

```python
# visualize attrition with respect to age
plt.figure(figsize=(10, 6))
sns.kdeplot(data, x='Age', hue='Attrition', fill=True)
plt.title('Attrition Distribution by Age')
plt.show()
```

```python
# visualize attrition with respect to gender
plt.figure(figsize=(8, 6))
sns.countplot(data, x='Gender', hue='Attrition')
plt.title('Attrition by Gender')
plt.show()
```

```python
# visualize attrition with respect to department
plt.figure(figsize=(8, 6))
sns.countplot(data, x='Department', hue='Attrition')
plt.title('Attrition by Department')
plt.show()
```

It can be observed that the highest attrition occurs in Research & Development Department

```python
# visualize attrition with respect to job role
plt.figure(figsize=(12, 6))
sns.countplot(data, x='JobRole', hue='Attrition')
plt.xticks(rotation=45)
plt.title('Attrition by Job Role')
plt.show()
```
